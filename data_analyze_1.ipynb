```python
TARGET_NAME = cred_default
BASE_FEATURE_NAMES = df_base.columns.drop(TARGET_NAME).tolist()
```

```python
def plot_feat_scatter(feat_name):
    fig, ax = plt.subplots(1, 2)
    ax1, ax2 = ax.flatten()
    
    print(f'{feat_name}:')
    
    ax1.scatter(df[cred_score], df[feat_name])
    ax2.hist(df[feat_name])
    
    fig.set_size_inches(12, 3)
    plt.subplots_adjust(wspace=0.3, hspace=0.3)

    plt.show()
```

```python
for col_name in BASE_FEATURE_NAMES:
    plot_feat_scatter(col_name)
```
